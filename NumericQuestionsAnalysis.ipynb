```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
everything = {} # this is the whole json object that contains all of the data

with open("data.json", "r") as inf:
    everything = json.load(inf)
```

```python
# A big huge block of code for slicing the data that probably could have been done better/cleaner/more efficiently with pandas
# but I didn't think of that right away so here it is

all_ratings = {}
male_ratings = {} # this will be male profs by dept
female_ratings = {} # this will be female profs by dept

only_females = [] # a list of tuples (course["rating_of_course"], course["rating_of_instruction"] , course["how_much_learned"], course["how_hard"])
only_males = [] # same as above

all_male_ratings = {} # ignores dept
all_female_ratings = {} # ignore dept


# initialize a bunch variable
all_male_avg_course_rating = 0 # the average course rating
all_male_avg_instructor_rating = 0 # the average instructor rating
all_male_avg_how_much_learned = 0 # the average how much learned
all_male_avg_how_hard = 0 # the average for how intellictually challenging
all_male_course_count = 0

all_female_avg_course_rating = 0
all_female_avg_instructor_rating = 0
all_female_avg_how_much_learned = 0
all_female_avg_how_hard = 0
all_female_course_count = 0

    
number_of_responses = 0
course_count = 0

total_comments = 0
male_total_comments = 0
female_total_comments = 0

for keys, values in everything["department"].items():
    print(keys)
    
    # variables for all of the course
    all_ratings[keys] = {}
    avg_course_rating = 0
    avg_instructor_rating = 0
    avg_how_much_learned = 0
    avg_how_hard = 0
    
    number_of_responses = 0
    course_count = 0
    
    # different set for men
    male_ratings[keys] = {}
    male_avg_course_rating = 0
    male_avg_instructor_rating = 0
    male_avg_how_much_learned = 0
    male_avg_how_hard = 0
    
    male_number_of_responses = 0
    male_course_count = 0
    
    # different set for women
    female_ratings[keys] = {}
    female_avg_course_rating = 0
    female_avg_instructor_rating = 0
    female_avg_how_much_learned = 0
    female_avg_how_hard = 0
    
    female_number_of_responses = 0
    female_course_count = 0
    
    # for each course in a dept
    for course in values:
        # this mini block is for the analysis by dept
        course_count += 1
        avg_course_rating += course["rating_of_course"]
        avg_instructor_rating += course["rating_of_instruction"] 
        avg_how_much_learned += course["how_much_learned"]
        avg_how_hard += course["how_hard"]
        
        number_of_responses += len(course["comments"])
        
        if course["instructor_gender"] == "F":
            female_course_count += 1
            female_avg_course_rating += course["rating_of_course"]
            female_avg_instructor_rating += course["rating_of_instruction"] 
            female_avg_how_much_learned += course["how_much_learned"]
            female_avg_how_hard += course["how_hard"]
        
            female_number_of_responses += len(course["comments"])
            
            all_female_avg_course_rating += course["rating_of_course"]
            all_female_avg_instructor_rating += course["rating_of_instruction"]
            all_female_avg_how_much_learned += course["how_much_learned"]
            all_female_avg_how_hard += course["how_hard"]
            
            all_female_course_count += 1
            
            only_females.append((course["rating_of_course"], course["rating_of_instruction"] , course["how_much_learned"], course["how_hard"]))
            
        elif course["instructor_gender"] == "M":
            male_course_count += 1
            male_avg_course_rating += course["rating_of_course"]
            male_avg_instructor_rating += course["rating_of_instruction"] 
            male_avg_how_much_learned += course["how_much_learned"]
            male_avg_how_hard += course["how_hard"]
        
            male_number_of_responses += len(course["comments"])
            
            all_male_avg_course_rating += course["rating_of_course"]
            all_male_avg_instructor_rating += course["rating_of_instruction"]
            all_male_avg_how_much_learned += course["how_much_learned"]
            all_male_avg_how_hard += course["how_hard"]
            
            all_male_course_count += 1
            only_males.append((course["rating_of_course"], course["rating_of_instruction"] , course["how_much_learned"], course["how_hard"]))

            
        else:
            print("Still no gender for {}".format(course["instructor_name"]))

    all_ratings[keys]["dept_course_rating"] = avg_course_rating/course_count
    all_ratings[keys]["dept_instructor_rating"] = avg_instructor_rating/course_count
    all_ratings[keys]["dept_how_much_learned"] = avg_how_much_learned/course_count
    all_ratings[keys]["dept_how_hard"] = avg_how_hard/course_count
    all_ratings[keys]["number_of_courses"] = course_count
    
    all_ratings[keys]["total_number_of_comments"] = number_of_responses
    total_comments += number_of_responses
    
    male_ratings[keys]["dept_course_rating"] = male_avg_course_rating/male_course_count
    male_ratings[keys]["dept_instructor_rating"] = male_avg_instructor_rating/male_course_count
    male_ratings[keys]["dept_how_much_learned"] = male_avg_how_much_learned/male_course_count
    male_ratings[keys]["dept_how_hard"] = male_avg_how_hard/male_course_count
    male_ratings[keys]["number_of_courses"] = male_course_count
    
    male_ratings[keys]["total_number_of_comments"] = male_number_of_responses
    male_total_comments += male_number_of_responses
    
    female_ratings[keys]["dept_course_rating"] = female_avg_course_rating/female_course_count
    female_ratings[keys]["dept_instructor_rating"] = female_avg_instructor_rating/female_course_count
    female_ratings[keys]["dept_how_much_learned"] = female_avg_how_much_learned/female_course_count
    female_ratings[keys]["dept_how_hard"] = female_avg_how_hard/female_course_count
    female_ratings[keys]["number_of_courses"] = female_course_count
    
    female_ratings[keys]["total_number_of_comments"] = female_number_of_responses
    female_total_comments += female_number_of_responses

# compute averages for the female and male 
all_female_ratings["overall_female_course_rating"] = all_female_avg_course_rating/female_course_count
all_female_ratings["overall_female_instructor_rating"] = all_female_avg_instructor_rating/female_course_count
all_female_ratings["overall_female_how_much_learned"] = all_female_avg_how_much_learned/female_course_count
all_female_ratings["overall_female_how_hard"] = all_female_avg_how_hard/female_course_count
all_female_ratings["number_of_course"] = all_female_course_count

all_male_ratings["overall_male_course_rating"] = all_male_avg_course_rating/male_course_count
all_male_ratings["overall_male_instructor_rating"] = all_male_avg_instructor_rating/male_course_count
all_male_ratings["overall_male_how_much_learned"] = all_male_avg_how_much_learned/male_course_count
all_male_ratings["overall_male_how_hard"] = all_male_avg_how_hard/male_course_count
all_male_ratings["number_of_course"] = all_male_course_count
```
